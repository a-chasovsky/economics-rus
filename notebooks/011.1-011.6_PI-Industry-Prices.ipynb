### 11.1. Средние цены производителей на отдельные виды промышленных товаров (с 1998 г.) - 1998-2009

*Источник: https://rosstat.gov.ru/storage/mediabank/Proizvoditeli_Cena-09.xlsx*  
*Единицы измерения: *

*Примечания: в трех датасетах для 1998-2009, 2010-2016, 2017-2022 годов названия товаров меняются, что делает затруднительными конкатенацию датафреймов. При необходимости данные из этих датасетов должны загружаться по-отдельности после сопоставления и выбора названий интересующего товара.*

*Например, "лесоматериалы круглые для выработки **пиломатериалов**, м3 плотн." в 1998-2009 году, "лесоматериалы круглые хвойных пород для **распиловки**, плотн м3" в 2010-2016 году, "лесоматериалы круглые хвойных пород для распиловки **и строгания**, плотн. м3." в 2017-2022 году.   
Отличия в 'хвойных' и 'пиломатериалов/распиловки'*  

```python
replace_dict_partial = {
    '1\\)': '',
    '2\\)': '',
    ',': '.'
}
replace_dict = {
    '...': np.NaN,
    '…': np.NaN,
    '..': np.NaN,
    '-': np.NaN,
    ' ': np.NaN
}
```

```python
pci_industry_link = \
    'https://rosstat.gov.ru/storage/mediabank/Proizvoditeli_Cena-11.xlsx'
```

```python
pci_industry2010_raw = pd.read_excel(pci_industry_link, sheet_name='1')
```

```python
pci_industry2010_raw
```

```python
pci_industry2010 = pci_industry2010_raw[2:-2].copy()
```

```python
pci_industry2010.iloc[0,0] = '-'
```

```python
# drop duplicate whitespaces
pci_industry2010.iloc[:, 0] = pci_industry2010.iloc[:, 0].apply(lambda x: remove_duplicated_whitespaces(x))
```

```python
pci_industry2010.columns = pci_industry2010.iloc[0, :]
```

```python
pci_industry2010 = pci_industry2010[1:].copy()
```

```python
pci_industry2010 = pci_industry2010.set_index('-', drop=True)
```

```python
pci_industry2010.index.name = None
pci_industry2010.columns.name = None
```

```python
pci_industry2010.index = [str(i).lower() for i in pci_industry2010.index]
```

```python
pci_industry2010 = pci_industry2010.T
```

```python
pci_industry2010 = pci_industry2010.replace(replace_dict_partial, regex=True)
```

```python
pci_industry2010 = pci_industry2010.replace(replace_dict)
```

```python
pci_industry2010 = pci_industry2010.astype(float)
```

```python
pci_industry2010 = pci_industry2010.round(1)
```

```python
pci_industry2010
```
