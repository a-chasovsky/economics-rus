```python
idxs_agriculture_link = \
    'https://rosstat.gov.ru/storage/mediabank/ind_sx-12.xlsx'
```

```python
idxs_agriculture_raw = pd.read_excel(idxs_agriculture_link, sheet_name='1')
```

```python
idxs_agriculture_raw
```

```python
idxs_agriculture = idxs_agriculture_raw[2:-2].copy()
```

```python
idxs_agriculture.iloc[0, 0] = 'idx'
```

```python
idxs_agriculture.loc[16, 'Unnamed: 0'] = 'Год'
```

```python
# drop duplicate whitespaces
idxs_agriculture.iloc[:, 0] = \
    idxs_agriculture.iloc[:, 0].map(remove_duplicated_whitespaces)
```

```python
idxs_agriculture.columns = idxs_agriculture.iloc[0, :]
```

```python
idxs_agriculture = idxs_agriculture[2:].copy()
```

```python
idxs_agriculture.columns = [str(i)[:4] for i in idxs_agriculture.columns]
```

```python
idxs_agriculture.columns = [to_int(i) for i in idxs_agriculture.columns]
```

```python
idxs_agriculture = idxs_agriculture.set_index('idx', drop=True)
```

```python
idxs_agriculture.index.name = None
```

```python
idxs_agriculture.index = [str(i).lower() for i in idxs_agriculture.index]
```

```python
idxs_agriculture = idxs_agriculture.T
```

```python
idxs_agriculture = idxs_agriculture.astype(float)
```

```python
idxs_agriculture = idxs_agriculture.round(1)
```
