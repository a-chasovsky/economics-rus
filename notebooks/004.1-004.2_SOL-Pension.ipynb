```python
replace_dict = {
    'г. Москва': 'Москва',
    'г. Санкт-Петербург': 'Санкт-Петербург',
    'г. Севастополь': 'Севастополь',
    'Кемеровская область': 'Кемеровская область - Кузбасс',
    'Республика Северная Осетия - Алания': 'Республика Северная Осетия-Алания'
}
pension_avg_2023 = pension_avg_2023.replace({'Регион': replace_dict})
```

```python
# list with all columns, except 'Регион'
columns_lvls = pension_avg_2023.columns[1:]
# list with all columns level0, except 'Регион'
columns_lvl0 = [i for i in pension_avg_2023.columns.droplevel(1)
                if i not in ['Регион']]
columns_lvl0 = list(set(columns_lvl0))
```

```python
# stay list1
date_list4 = ['1 августа', '1 января']
# first: remove all months except '1 августа' and '1 января'
# for every column in level0 columns
for col_lvl0 in columns_lvl0:
    # if level1 columns frequency more than 1
    columns_lvl1 = pension_avg_2023[col_lvl0].columns
    if len(columns_lvl1) > 1:
        # drop columns, if level1 column not in 'date_list3'
        columns_to_drop = [
            (col_lvl0, col_lvl1) for col_lvl1 in columns_lvl1
            if col_lvl1 not in date_list4]  
        pension_avg_2023 = pension_avg_2023.drop(columns=columns_to_drop, axis=1)
```
