# Section I. Основные показатели и Население (Summary, Population)

## 1. Основные макроэкономические показатели. Оперативная информация

```python
macro_primary_link = 'http://sophist.hse.ru/hse/1/tables/BRDATA.htm'
```

```python
macro_primary = pd.read_html(
    io=macro_primary_link, flavor='html5lib', thousands='.',
    decimal=',', encoding='CP1251', header=0)[0]
```

```python
macro_primary
```

```python
macro_primary = macro_primary.rename(
    columns={'Предыдущее значение': 'Предыдущий период'})
macro_primary = macro_primary.rename(
    columns={'Последнее значение': 'Последний период'})
```

```python
for i in macro_primary.index:
    macro_primary.loc[i, 'Предыдущее значение'] = \
        macro_primary.loc[i, 'Предыдущий период'].split()[0]
    macro_primary.loc[i, 'Предыдущий период'] = \
        ' '.join(macro_primary.loc[i, 'Предыдущий период'].split()[1:])
    macro_primary.loc[i, 'Последнее значение'] = \
        macro_primary.loc[i, 'Последний период'].split()[0]
    macro_primary.loc[i, 'Последний период'] = \
        ' '.join(macro_primary.loc[i, 'Последний период'].split()[1:])
```

```python
new_cols_order = [
    'Название показателя', 'Код показателя', 'Единицы измерения',
    'Предыдущий период', 'Предыдущее значение',
    'Последний период', 'Последнее значение', 'Комментарий'
]
```

```python
macro_primary = macro_primary[new_cols_order]
```
