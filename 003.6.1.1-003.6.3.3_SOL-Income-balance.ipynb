#### 3.6.1. <font color='#2A5899'>Income and Spendings: Nominal</font>

##### Data

```python
income_spendings_nom_link = \
    ('data/02-Population/02-Standard-of-living/01-Income/'
     'urov_17g-nm.xlsx')
```

```python
income_spendings_nom_raw = pd.read_excel(income_spendings_nom_link, sheet_name='номинал')
```

```python
income_spendings_nom_raw.head(10)
```

```python
income_spendings_nom_raw.tail(10)
```

```python
income_spendings_nom = income_spendings_nom_raw.copy()
```

```python
# cut rows
income_spendings_nom = transform_cut_rows(income_spendings_nom, 2, 2)
```

```python
income_spendings_nom = income_spendings_nom.rename(
    columns={income_spendings_nom.columns[0]: 'index2'})
```

```python
income_spendings_nom['index0'] = income_spendings_nom['index2'].copy()
income_spendings_nom['index1'] = income_spendings_nom['index2'].copy()
```

##### Multiindex level0

```python
income_spendings_nom['index0'] = \
    income_spendings_nom['index0'].apply(lambda x: x.replace(' ', ''))
income_spendings_nom['index0'] = \
    income_spendings_nom['index0'].apply(lambda x: x.replace(':', ''))
```

```python
keep_list = [
    'ДОХОДЫ', 'РАСХОДЫ', 'СБЕРЕЖЕНИЯ', 'Справочно']
income_spendings_nom['index0'] = income_spendings_nom['index0'].apply(
    lambda x: replace_inlist(x, keep_list, np.NaN, kind='not'))
```

```python
income_spendings_nom = transform_fill_values_by_previous(
    income_spendings_nom, kind='column', column_name='index0')
```

```python
income_spendings_nom['index0'] = income_spendings_nom['index0'].str.capitalize()
```

##### Multiindex level1

```python
income_spendings_nom['index1'] = \
    income_spendings_nom['index1'].apply(lambda x: x.replace(':', ''))
```

```python
keep_list = [
    'I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'Справочно']
income_spendings_nom['index1'] = income_spendings_nom['index1'].apply(
    lambda x: transform_string_replace_value_if_not_contains(
        x, keep_list, np.NaN, full_list=False))
```

```python
income_spendings_nom['index2'] = income_spendings_nom['index2'].apply(
    lambda x: transform_string_replace_value_if_contains(
        x, keep_list, 'Всего', full_list=False))
```

```python
income_spendings_nom = transform_fill_values_by_previous(
    income_spendings_nom, kind='column', column_name='index1')
```

##### Replace some values

```python
income_spendings_nom.loc[2] = income_spendings_nom.loc[2].apply(
    lambda x: transform_replace(x, ' год', '')).copy()
```

```python
income_spendings_nom['Unnamed: 1'] = income_spendings_nom['Unnamed: 1'].apply(
    lambda x: transform_replace(x, 'х', '-'))
```

##### Multiindex Create

```python
# columns to multiindex
income_spendings_nom = transform_make_index_from_columns(
    income_spendings_nom, ['index0', 'index1', 'index2'])
```

```python
# drop columns, that became multiindexes
income_spendings_nom = \
    income_spendings_nom.drop(['index0', 'index1', 'index2'], axis=1)
```

```python
columns_names = income_spendings_nom.iloc[0, :].tolist()
columns_names = [i.replace(' ', '') for i in columns_names]
income_spendings_nom.columns = columns_names.copy()
income_spendings_nom = \
    income_spendings_nom.drop(income_spendings_nom.iloc[0].name, axis=0)
```

##### Drop NaNs and round values if possible

```python
income_spendings_nom = income_spendings_nom.dropna()
```

```python
income_spendings_nom = income_spendings_nom.applymap(transform_round)
```

##### Final Datasets

```python
income_nominal = income_spendings_nom.loc['Доходы'].copy()
```

```python
spendings_nominal = income_spendings_nom.loc['Расходы'].copy()
```

```python
savings_nominal = income_spendings_nom.loc['Сбережения'].copy()
savings_nominal.index = savings_nominal.index.droplevel(1)
```

```python
isna(income_nominal)
```

```python
income_nominal
```

```python
isna(spendings_nominal)
```

```python
spendings_nominal
```

```python
isna(savings_nominal)
```

```python
savings_nominal
```

```python
saveit_excel(
    data=spendings_nominal,
    filename=dir,
    path=save_excel,
    sheet='Расходы номинал'
)
```
